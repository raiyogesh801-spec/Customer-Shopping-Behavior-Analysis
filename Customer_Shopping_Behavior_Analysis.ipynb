```python
df.isnull().sum() # Checking if my Dataset Having Missing Values or Not.
# So Review rating is Contaning Missing values.
```

```python
## Now filling Missing values I Can use  Mean and Median But Mean Can get Influenced By Outliers
# So I'm using Median Here.
```

```python
# Im using Median vaulue of review rating of Each Category so that Missing values of any get replace by its own Catgeory.
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum() # Lets now check for missing values 
```

```python
# Now we can we columns having Spaces and Upper and Lower Cases and this Can through error in futures 
# So I will Chnaged Columns Names In Snak-Casing (lowerCases + _)
```

```python
df.columns = df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns = {'purchase_amount_(usd)':'purchased_amount'})
```

```python
df.columns
```

```python
#  So I'll Create a column age_group to group Customers into 4 four groups:
labels = ['young adult','adult','middle-aged', 'senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels = labels) # qcut splits ages into 4 equal size groups based on the Data distrubtion 
## and assinged the lables we difined 
```

```python
df[['age','age_group']].head(10)
```

```python
# let's create another column purchase_frequency_days Becuase The column we already have in our dataset 
# is in Text form and working with text Make anylysis harder.

frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly': 30,
    'Quarterly' : 90,
    'Bi-weekly': 14,
    'Annually': 365,
    'Every 3 Months' : 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
